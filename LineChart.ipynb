```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.widgets import CheckButtons
import numpy as np
```

```python
df = pd.read_excel('scores.xlsx', 'Sheet1')
df
```

```python
freq_table = df[['freecharge_score', 'mobikwik_score', 'gpay_score', 'paypal_score', 'paytm_score', 'phonepe_score']].apply(pd.Series.value_counts)
```
